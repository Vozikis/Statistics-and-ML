```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## create data for the histogram

# number of data points
n = 1000

# generate data - log-normal distribution
data = np.exp( np.random.randn(n)/2 )
plt.plot(data,'s')
#no values under 0 and not many values close to 0
```

```python
# show as a histogram

# number of histogram bins
k = 40

plt.hist(data,bins=k)
plt.show()
```

```python
# another option
#what if we want to get those values back!
y,x = np.histogram(data,bins=k)

# bin centers
#because we have 40 y values and 41 x values due to net getting the middle of each pf the bins
xx = (x[1:]+x[:-1])/2

plt.plot(xx,y)
plt.show()
#here we get the same data representation but instead of bars we have lines
```

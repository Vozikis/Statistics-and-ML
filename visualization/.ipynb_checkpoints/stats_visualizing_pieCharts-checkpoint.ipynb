```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
```

```python
## create data for the plot

nbins = 5#number of bins
totalN = 100#number of datapoints

rawdata = np.ceil(np.logspace(np.log10(1/2),np.log10(nbins-.01),totalN))
#we add this nbins-.01 because due to computer rounding errors we might get a value a bit bigger than 5. like
#5.0000000000000032 which ceil will automatically convert to 6. We dont want that to happen
#creating logarithmic spaced numbers going from 1/2 until the bin-0.1
#numbers start from 0.5 and keep on getting larged and larger until we get to arround 5 which is the bin number
#we also include ceil so we round up the numbers into the integer level 0.5 ->1 , 4.9 ->5 etc
#fewer 5s than 4s and fewer 4s than 3 etc

# prepare data for pie chart
uniquenums = np.unique(rawdata)
data4pie = np.zeros(len(uniquenums))
#gives the number of times each number pops up
for i in range(len(uniquenums)):
    data4pie[i] = sum(rawdata==uniquenums[i])

print(data4pie)
```

```python
# show the pie chart
#input the data and for the labels we convert the values into percentage with 100*data4pie/sum(data4pie)
plt.pie(data4pie,labels=100*data4pie/sum(data4pie))
plt.show()
```

```python
# another option
plt.pie(data4pie,labels=['zero','one','two','three','four'],explode=[0,.1,0,.15,0])
plt.show()
```

```python
## for continuous data

# generate log-normal distribution
data = np.exp( np.random.randn(1000)/10 )


# generate bins using histogram
histout = np.histogram(data,bins=6)

# and show that as a pie chart
plt.pie(histout[0])
plt.show()
```

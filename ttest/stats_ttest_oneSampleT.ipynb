```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
```

```python
## generate the data

# parameters
N = 20  # sample size
popMu = .5 # true population mean
data  = np.random.randn(N) + popMu #gaussian distributed but with added random noise


# let's see what the data look(s) like
plt.plot(data,'ko',markerfacecolor='w',markersize=10)#we changed ko- to ko to remove the liens
plt.xlabel('Data index')
plt.ylabel('Data value')
plt.show()

### question: Should there be lines in this plot?
#there should be NO lines because we cant have data inbetween 2 points
```

```python
## "manual" t-test

# the null-hypothesis value
H0val = 0

# compute the t-value
t_num = np.mean(data) - H0val
t_den = np.std(data) / np.sqrt(N)
tval = t_num / t_den

# degrees of freedom
df = N-1 #as soon as we know the mean we only need n-1 degress of freedom
#because we can find the last result.

# p-value
pval = 1-stats.t.cdf(abs(tval),df)#percentage of those that are more than our t value
#what is the probability of getting a bigger value than the t given the null hypothesis

# show the H0 parameter distribution and observed t-value
x = np.linspace(-4,4,1001)
tdist = stats.t.pdf(x,df) * np.mean(np.diff(x))

plt.plot(x,tdist,linewidth=2)
plt.plot([tval,tval],[0,max(tdist)],'r--')
plt.legend(('H_0 distribution','Observed t-value'))
plt.xlabel('t-value')
plt.ylabel('pdf(t)')
plt.title('t(%g) = %g, p=%g'%(df,tval,pval))
plt.show()
```

```python
## now using the Python function

t,p = stats.ttest_1samp(data,H0val)#computes two tailed, although we are only considering one tail


print(t,p)
# do these values match our manually computed values?
#difference in values due to degrees of freedom. the latter uses n degress of freedom!
```
